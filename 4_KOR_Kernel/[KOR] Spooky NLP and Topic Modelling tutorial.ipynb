# 2. 자연어 처리

거의 모든 자연어 처리(컴퓨터와 인간 언어 사이의 상호 작용을 탐구하는 분야) 작업(토픽 모델링, 단어 클러스터링 또는 document-text classification 등)에서 input raw text를 모델과 기계가 읽을 수 있는 형식으로 변환하려면 일반적으로 항상 몇 가지 전처리 단계를 거쳐야합니다. 확실하게 RandomForest 모델에 단어 단락을 제공하고 해당 단락의 작가를 즉시 예측할 수는 없습니다. 그 뒤에서, 텍스트 전처리는 다음과 같은 몇 가지 간단한 단계로 요약될 수 있습니다:

1. **Tokenization** - 텍스트를 개별 단어로 분리합니다.<br>
2. **Stopwords** - 발생 빈도는 관련 텍스트 감지에 도움이 되지 않으므로 너무 자주 발생하는 단어는 버리세요. (또한, 아주 드물게 발생하는 단어들도 버립니다.)<br>
3. **Stemming** - 변형된 단어들을 하나의 부모 단어로 결합하여 여전히 동일한 의미를 전달할 수 있게 합니다.<br>
4. **Vectorization** - 텍스트를 벡터 형식으로 변환합니다. 가장 간단한 방법 중 하나는 유명한 단어별 접근 방식(bag-of-words approach)입니다. 이 방법을 사용하면 각 문서 또는 텍스트의 말뭉치에 대한 행렬을 만들 수 있습니다. 가장 간단한 형태로, 이 행렬은 단어 빈도(단어 수)를 저장하며 원시 텍스트의 벡터화라고도 합니다.<br>


**Natural Language Toolkit (NLTK)** : 자연어 처리를 보다 편리하게 하기 위해 NLP에서 가장 편리한 툴킷 중 하나인 자연어 툴킷(NLTK 모듈이라고도 함)을 소개하겠습니다. 툴킷을 가져오려면, 다음과 같이 쉽습니다:

```python
import nltk
```

## 2a. Tokenization

토큰화의 개념은 주어진 문서에서 일련의 문자(Python strings을 생각)를 가져와서 이 방법의 시조격 '토큰'인 개별 구성 부분으로 나누는 것입니다. 그것들을 대충 문장에서의 단수형 단어로 생각할 수도 있습니다. 인수에서의 식별자를 기반으로 하여 파이썬 리스트로 분리하는 문자열에서 split() 메서드를 대충 구현할 수도 있습니다. 실제로는 그렇게 사소한 것이 아닙니다.

여기서 우리는 훈련 데이터의 텍스트의 첫 문장을 다음과 같이 공백으로 나눕니다:

```python
# Storing the first text element as a string
first_text = train.text.values[0]
print(first_text)
print('='*90)
print(first_text.split(' '))
```

그러나 첫번째 토큰화 시도에서 볼 수 있듯이, 문장을 개별 요소(또는 용어)로 분리하는 것이 완전히 정확하지는 않습니다. 예를 들어, 'process'라는 용어가 포함된 리스트의 두번째 요소를 보세요. 이상적으로는 쉼표와 단어가 리스트 내에서 서로 다른 두 개의 개별 요소로 되어있기를 원합니다. 순수한 파이썬 리스트 작업을 수행하는 것은 상당히 복잡하므로 NLTK 라이브러리가 그 역할을 할 것입니다. 'word_tokenize()' (TreebankWord tokenizer) 이라는 편리한 방법은 다음과 같이 단수 단어들과 구두점들을 개별 요소로 자동 제거합니다:

```python
first_text_list = nltk.word_tokenize(first_text)
print(first_text_list)
```

## 2b. Stopword Removal

위의 불용어는 일반적으로 말뭉치에서 자주 나타나고 높은 빈도수를 갖는 단어들이기 때문에 학습 모델이 다른 텍스트들과 구별하지 못하게 되므로 학습 또는 예측 과정에 기여하지 않습니다. 불용어에는 'to' 나 'the'와 같은 용어들이 포함되므로 전처리 단계에서 제거하는 것이 좋습니다. 편리하게도, NLTK에는 153 개의 영어 불용어들이 사전 정의된 리스트가 제공됩니다.

```python
stopwords = nltk.corpus.stopwords.words('english')
len(stopwords)
```

## 2c. Stemming and Lemmatization

불용어들을 제거한 이후, NLP의 다음 단계는 Stemming(어간 추출) 입니다. 이 단계에서의 작업은 유사한 단어의 다양한 변형을 단일 용어로 축소(여러 가지들을 하나의 줄기 단어로)하는 시도를 하는 것입니다. 따라서 우리가 'running', 'runs', 'run'을 갖고 있다면 이 세 개 단어들이 'run' 이라는 단어로 되기를 원할 것입니다(물론 과거, 현재 또는 미래 시제의 세분성을 잃게 되지만).

NLTK는 포터(Porter) 스테밍 알고리즘, 랭커스터(lancaster) 스테머, 스노우볼 스테머와 같은 변형 형태를 포함한 다양한 스테머들을 제공할 수 있습니다. 다음 예제에서는 포터 스테머 인스턴스를 다음과 같이 만들 것입니다:

```python
stemmer = nltk.stem.PorterStemmer()
```

이제 우리는 스테머를 이용해서 테스트 단어들('running', 'runs', 'run')을 하나의 줄기 단어로 줄일 수 있는지 볼 것입니다. 편리하게 다음과 같이 스테머를 테스트할 수 있습니다: 

```python
print('The stemmed form of running is: {}'.format(stemmer.stem('running')))
print('The stemmed form of runs is: {}'.format(stemmer.stem('runs')))
print('The stemmed form of run is: {}'.format(stemmer.stem('run')))
```

보시다시피, 스테머는 위에서 주어진 단어들을 성공적으로 기본 형태로 축소했으며 이후의 학습 및 분류 작업을 할 때 단어의 데이터셋 크기를 줄이는 데 가장 도움이 될 것입니다.

그러나 스테밍에 한 가지 결함이 잇는데 이는 특정 단어를 인간이 인식할 수 있는 기본 형식으로 줄이기 위해 단어들의 끝을 잘라내는 과정에 상당한 휴리스틱이 포함되어 있다는 사실입니다. 따라서 이 과정에서는 다음과 같이 단어를 축소할 때 어휘나 단어 형식을 고려하지 않습니다:

```python
print('The stemmed form of leaves is: {}'.format(stemmer.stem('leaves')))
```


따라서 우리는 어간 추출 대신에 사용할 수 있는 다른 것으로 바꿉니다. 이 방법을 lemmatization(표제어 추출)이라고 하며 이전 방법과 동일한 효과를 얻는 것을 목표로 합니다. 그러나 스테머와는 달리 데이터셋을 lemmatizing하는 것은 실제 사전이나 어휘(Lemma)를 기반으로 단어를 줄이는 것을 목표로 하기 때문에 어휘적 의미를 갖지 않는 형태로 단어를 잘라내지 않습니다. 여기서 우리는 NLTK를 다시 사용하여 lemmatizer(WordNet variant)를 초기화하고 다음과 같이 단어가 어떻게 축소되는지 검사할 수 있습니다:

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
print('The lemmatized from of leaves is: {}'.format(lemm.lemmatize('leaves')))
```

훌륭합니다. 이제 lemmatizer가 단어를 훨씬 어휘적으로 이해하는 형태로 작동하고 단어들을 분해함을 알 수 있습니다.

## 2d.Vectorizing Raw Text
NLP 문헌의 방대한 컬렉션에는 원시 텍스트를 분석하기 위한 여러 가지 목적이 있습니다. 경우에 따라 한 텍스트의 본문과 다른 본문의 유사성을 비교하려는 경우(클러스터링 기법/거리 측정), 텍스트 분류(이 competition의 목적), 텍스트의 본문을 이루는 주제를 밝히는 경우(이 notebook의 목적)가 있습니다. 우리가 염두에 두고 있는 주제를 밝히기 위해 이제 원문을 학습 모델에 주는 방법을 생각해야 합니다. 토큰화, 불용어 제거 및 형태소 분석(또는 표제어 추출)에 대해 이미 다루었으므로 이제는 상당히 깨끗한 텍스트 데이터셋에 도달했습니다. 그러나 이 시점에서, 우리의 원시 텍스트는 인간이 읽을 수는 있지만 불행히도 기계에서는 아직 읽지 못합니다. 기계는 비트와 숫자를 읽을 수 있으므로, 매우 일반적인 기법인 Bag-of-Words(BoW)를 이용하기 위해 먼저 텍스트를 숫자로 변환해야합니다.

**The Bag of Words approach**
이 기법은 단어 수를 시작 블럭으로 사용하고 (전체 텍스트에서) 각 단어의 발생을 해당 단어의 특정한 벡터에 기록합니다. 예를 들어, 'I love to eat Burgers', 'I love to eat Fries' 라는 두 문장이 주어지면 먼저 [I, love, to, eat, Burgers, Fries] 단어 수로부터 6개 단어를 얻기 위해 토큰화를 해야합니다.

Bag of Words 접근 방식을 통해 텍스트를 벡터화하면 각 단어마다 하나씩 6개의 고유한 벡터가 생성됩니다. 따라서 열(또는 feature)들을 이루는 (텍스트 대신에) 숫자들로 이뤄진 행을 갖게 됩니다. 각 단어는 이제 새롭게 변형된 데이터셋에서 개별 feature/열 이 되었습니다. 이 점을 설명하기 위해 Scikit-learn 라이브러리를 사용하여 다음과 같이 CountVectorizer 메서드를 통해 단어 개수(용어 빈도)의 벡터를 생성하는 vectorizer를 구현합니다.

* 1. Latent Dirichlet Allocation - 말뭉치의 단어에 가중치를 할당하여 데이터셋에 잠재된 주제를 발견하는 확률론적 생성 모델로, 각 주제는 각 단어에 서로 다른 확률 가중치를 할당합니다.<br>
* 2. Non-negative Matrix Factorization - input 행렬을 가져와서 이 행렬을 두 개의 다른 행렬로 인수 분해(factorization)하는 근사법으로, 행렬의 값이 음이 아니어야 합니다. 

```python
# Define helper function to print top words
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = '\nTopic #{}'.format(index)
        message += ' '.join([feature_names[i] for i in topic.argsort()[:-n_top_words -1 :-1]])
        print(message)
        print('='*70)
```

감사하게도, 우리는 그럴 필요가 없습니다. Sklearn vectorizer에 대한 주요 세부 사항은 생략했지만 이 시점에서 언급할 것입니다. CountVectorizer로 원시 텍스트를 벡터화하면, 토큰화 및 불용어 필터링의 이중 단계들이 자동으로 상위 레벨 구성 요소로 포함됩니다. 이전에 섹션 2a에서 소개한 NLTK 토크나이저와 달리 Sklearn의 토크나이저는 단일 문자 용어들('a', 'w' 등)을 버리고 기본적으로 모든 단어를 소문자로 합니다. Sklearn에서 불용어를 필터링하는 것은 내장된 영어 불용어 리스트가 자동으로 사용되는 'stop_words' 인수에 'english' 값을 전달하는 것만큼 편리합니다.

불행히도, vectorizer에는 내장된 lemmatizer가 없으므로 몇 가지 옵션이 있습니다. 벡터화를 위해 데이터를 공급하기 전에 매번 따로 구현하거나 이 기능을 포함하도록 sklearn 구현을 확장하세요. 운이 좋게도, 우리는 다음과 같은 'build_analyzer' 메서드를 덮어써서 CountVectorizer 클래스를 확장할 수 있는 후자의 옵션이 있습니다.

**Extending the CountVectorizer class with a lemmatizer**

```python
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
```

여기에서는 객체 지향 프로그래밍(OOP)의 미묘한 개념을 활용했습니다. 우리는 본질적으로 원본 Sklearn의 CountVectorizer 클래스를 상속하고 서브클래싱했으며 원시 텍스트 행렬의 각 리스트에 대해 lemmatizer를 구현하여 build_analyzer 메서드를 덮어썼습니다.

```python
# Storing the entire training text in a list
text = list(train.text.values)
# Calling our overwrite Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df=0.95,
                                     min_df=2,
                                     stop_words='english',
                                     decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
```

**Revisiting our Term frequencies**<br>

lemmatized count vectorizer를 구현한 후 (빈도 수에 따른) 상위 50개 단어의 빈도를 플롯으로 다시 살펴보겠습니다. 플롯에서 볼 수 있듯이, 우리의 모든 전처리 노력은 헛된 것이 아니었습니다. 불용어를 제거하면 나머지 단어들이 이전의 용어 빈도 플롯의 불용어를 볼 수 있는 곳에서 더 의미있게 보입니다.

마지막으로 토픽 모델링과 두 개의 비지도 학습 알고리즘 구현에 대해 다룹니다. 첫 번째 방법은 Latent Dirichlet Allocation입니다. 이 LDA 알고리즘에는 두 가지의 다른 구현방법이 있지만 이 노트북에서는 Sklearn의 구현을 사용할 겁니다. 잘 알려진 또다른 LDA구현은 Radium Rehurek의 [gensim](https://radimrehurek.com/gensim/)이므로 확인해보세요.

LDA 알고리즘은 먼저 주제의 혼합 모델을 통해 문서를 모델링합니다. 이 주제에서, 단어들은 이러한 주제의 확률 분포를 기반으로 가중치를 할당받습니다. LDA 사용자가 특정 단어가 주제에 속할 가능성을 말할 수 있도록 하는 것은 단어에 대한 확률론적 할당입니다. 결과론적으로 특정 주제에 할당된 단어 모음을 통해 해당 주제가 어휘 관점에서 실제로 무엇을 나타낼 수 있는지를 얻을 수 있습니다.

표준 LDA 모델에는 모델을 호출하기 전에 염두에 두고 프로그래밍 방식으로 조정해야하는 몇 가지 주요 매개변수(key parameter)들이 있습니다:

* 1. n_components: 모델에 지정하는 토픽 수<br>
* 2. α parameter : 이전 문서 토픽에 연결될 수있는 dirichlet 매개변수<br>
* 3. β parameter : 이전 토픽 단어에 연결된 dirichlet 매개변수

알고리즘을 호출하기 위해서, Sklearn의 LatentDirichletAllocation 함수를 통해 LDA 인스턴스를 생성합니다. 다양한 매개변수들은 어떤 종류의 유효성 검사 체계를 통해 이상적으로 얻을 수 있습니다. 이 경우, n_components(또는 topic number)의 최적값은 Kmeans + Latent Semantic Analysis Scheme을 수행하여 발견하였습니다. 이에 따라 Kmeans 군집 수와 LSA 차원 수를 반복하여 가장 우수한 실루엣 평균 점수를 냈습니다.

```python
lda = LatentDirichletAllocation(n_components=11, max_iter=5, 
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```
